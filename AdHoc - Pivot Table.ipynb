```python
import json
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('transaction-data-adhoc-analysis.json','r') as f:
    data = json.load(f)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['transaction_items'].str.split(';')
df['transaction_items'] = df['transaction_items'].map(str)
```

```python
df['transaction_items'] = df['transaction_items'].str.split(';')
```

```python
df
```

```python
# new data frame with new granularity
df2 = df.explode('transaction_items')
```

```python
df2
```

```python
# parse into transaction date
df2['transaction_date'] = pd.to_datetime(df2['transaction_date'])
```

```python
df2
```

```python
# get the number of items
df2['Quantity'] = df2['transaction_items'].str[-2:]
```

```python
df2
```

```python
df2["Quantity"] = df2["Quantity"].str.replace(")","")
```

```python
df2
```

```python
# clean up
df2.loc[df2['transaction_items'].str.contains('Beef Chicharon'), ['items']]='Beef Chicharon'
df2.loc[df2['transaction_items'].str.contains('Kimchi Seaweed'), ['items']]='Kimchi Seaweed'
df2.loc[df2['transaction_items'].str.contains('Nutrional Milk'), ['items']]='Nutrional Milk'
df2.loc[df2['transaction_items'].str.contains('Gummy Vitamins'), ['items']]='Gummy Vitamins'
df2.loc[df2['transaction_items'].str.contains('Yummy Vegetables'), ['items']]='Yummy Vegetables'
df2.loc[df2['transaction_items'].str.contains('Orange Beans'), ['items']]='Orange Beans'
df2.loc[df2['transaction_items'].str.contains('Gummy Worms'), ['items']]='Gummy Worms'
```

```python
# convert to integer to perform aggregate
df2['Quantity']=df2['Quantity'].astype(int)
```

```python
# convert to months
df2['Month'] = df2['transaction_date'].dt.month

df2
```

```python
# convert to months
df2.loc[df2['Month']==1, ['Month']]='January'
df2.loc[df2['Month']==2, ['Month']]='February'
df2.loc[df2['Month']==3, ['Month']]='March'
df2.loc[df2['Month']==4, ['Month']]='April' 
df2.loc[df2['Month']==5, ['Month']]='May'
df2.loc[df2['Month']==6, ['Month']]='June' 
```

```python
df2
```

```python
# pivot table that shows the breakdown of the count of each item sold per month
df2.pivot_table(index = "items", columns = "Month", values = 'Quantity', aggfunc='sum')
```

```python
# pivot table that shows the total sales for the given item in the given month
df2.pivot_table(index = "items", columns = "Month", values = 'transaction_value', aggfunc='sum')
```

```python
df3 = df2.pivot_table(index = "items", columns = "Month", values = 'transaction_value', aggfunc='sum')
```
